```python
# 정규표현식 사용

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

res = urlopen('https://davelee-fun.github.io/blog/crawl_test_css.html')
soup = BeautifulSoup(res, 'html.parser') 

data = soup.select('ul#dev_course_list li.course')
for item in data:
    print (re.sub('\[[0-9]+\]', '', item.get_text())) # 숫자 제거 
    
# \[[0-9]+\] : [0-9]는 0부터 9까지의 숫자 중 하나를 의미하고, +는 1개 이상의 숫자가 나올 수 있다는 의미이다.
```

```python
import openpyxl
import re

regex = re.compile('  [A-Za-z] +\.') # 정규표현식을 사용하여 영어만 추출
work_book = openpyxl.load_workbook('train.xlsx')

# active sheet
work_sheet = work_book.active

# work_sheet.rows : 모든 행을 가져옴
for each_row in work_sheet.rows:
    print (each_row[3].value)
    print (regex.findall(each_row[3].value)) 
    
work_book.close()
```

```python
import re
string = "(Grace)"
re.sub('[^A-Za-z0-9]', '', string) # 문자, 숫자 아닌 데이터 찾아서 ''로 대체
```

```python
string.replace("(",'').replace(")",'')  # 문자열에서 특정 문자 제거 "
```

```python
# 정규 표현식 라이브러리 임포트
import re

# 정규 표현식 패턴 생성
pattern = re.compile('D.A')
```

```python
# 반복 ?, *, +

# ? : 0회 혹은 1회 반복
# * : 0회 이상 반복
# + : 1회 이상 반복

pattern = re.compile('D?A') # D가 0회 혹은 1회 반복되고 A가 1회 반복되는 패턴
```

```python
pattern.search('DA')
```

```python
pattern.search('A')
```

```python
pattern.search('ADA')
```

```python
pattern = re.compile('D*A') # D가 0회 이상 반복되고 A가 1회 반복되는 패턴
```

```python
pattern.search('DA')
```

```python
pattern.search('DDDDDDDDDDDDDDDA')
```

```python
pattern.search('AD*A')
```

```python
pattern.search('ADDDDDDDDDDDDDDA')
```

```python
pattern = re.compile('D+A') # D가 1회 이상 반복되고 A가 1회 반복되는 패턴
```

```python
pattern.search('DA') 
```

```python
pattern.search('A') 
```

```python
# {n} : n회 반복
# {n, m} : n회 이상 m회 이하 반복

pattern = re.compile('AD{2}A') # D가 2회 반복되고 A가 1회 반복되는 패턴
```

```python
pattern.search('ADA')
```

```python
pattern.search('ADDA')
```

```python
pattern = re.compile('AD{2,4}A') # D가 2회 이상 4회 이하 반복되고 A가 1회 반복되는 패턴
```

```python
pattern.search('ADA')
```

```python
pattern.search('ADDA')
```

```python
# [ ] : 문자 집합

pattern = re.compile('[ABC]') # A, B, C 중 하나가 있는 패턴
```

```python
pattern.search('A123')
```

```python
# (-) : 범위

pattern = re.compile('[A-Z]') # A부터 Z까지 중 하나가 있는 패턴
```

```python
pattern.search('a123')
```

```python
pattern.search('A123')
```

```python
# [a-zA-Z0-9] : 모든 알파벳과 숫자

pattern = re.compile('[a-zA-Z0-9]') # 모든 알파벳과 숫자 중 하나가 있는 패턴
```

```python
pattern.search('123')
```

```python
# [] 안에서 ^ 사용시 그 뒤에 오는 문자가 아닌 패턴을 찾음

pattern = re.compile('[^A-Z]') # 대문자가 아닌 패턴
```

```python
pattern.search('a123')
```

```python
# [가-힣] : 모든 한글

pattern = re.compile('[가-힣]') # 모든 한글 중 하나가 있는 패턴
```

```python
pattern.search('가나다')
```

```python
# match : 문자열의 처음부터 정규식과 매칭되는지 확인
# search : 문자열 전체에서 정규식과 매칭되는지 확인

pattern = re.compile('D.A')
```

```python
match = pattern.match('DAA') # 문자열의 처음부터 매칭되는지 확인
```

```python
searched = pattern.search('DAA') # 문자열 전체에서 매칭되는지 확인
```

```python
print(match)
```

```python
print(searched)
```

```python
#findall : 정규식과 매칭되는 모든 문자열을 리스트로 반환

import re
pattern = re.compile('[a-z]+')
findalled = pattern.findall('Hello World')
```

```python
print(findalled)
```

```python
pattern2 = re.compile('[A-Za-z]+')
```

```python
findalled2 = pattern2.findall('Hello World')
```

```python
print(findalled2)
```

```python
import re
pattern = re.compile('[a-z]+')
findalled = pattern.findall('Hello World')
if len(findalled) > 0:
    print(findalled)
else:
    print('매칭되는 문자열이 없습니다.')
```

```python
# split : 정규식과 매칭되는 문자열을 기준으로 문자열을 나누어 리스트로 반환

import re
pattern2 = re.compile(':')
splited = pattern2.split('apple:orange:banana')
```

```python
print(splited)
```

```python
import re
pattern2 = re.compile(' [A-Z]{2} ') # 띄어쓰기로 구분된 두 글자의 대문자만 추출
splited = pattern2.split('apple US banana UK orange')
print(splited)
```

```python
# sub : 정규식과 매칭되는 문자열을 다른 문자열로 대체

import re
pattern = re.compile('-')
subed = pattern.sub('/', '2025-01-01')
```

```python
print(subed)
```

```python
import openpyxl
work_book = openpyxl.load_workbook('data_kr.xlsx')
work_sheet = work_book.active
for each_row in work_sheet.rows:
    print(re.sub('-[0-9]{7}', '-*******', each_row[1].value)) # -[0-9]{7} : -와 7개의 숫자가 매칭되는 패턴을 찾아서 -*******로 대체
work_book.close()
```
