````python
# naver Open API 사용하기

import requests
import openpyxl

client_id = 'twf9t32hX6dHhrzVza3t'
client_secret = '5cpTnSNQMG'

start = 1
num = 0

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['번호', '제목', '링크'])

for index in range(10):
    start_number = start + (index * 100)
    naver_open_api = 'https://openapi.naver.com/v1/search/news.json?query=python&display=100&start=' + str(start_number)
    header_params = {"X-Naver-Client-ID": client_id, "X-Naver-Client-Secret": client_secret}
    res = requests.get(naver_open_api, headers = header_params)
    if res.status_code == 200:
        data = res.json()
        for index, item in enumerate(data['items']): # enumerate 함수를 사용하면 인덱스 번호와 컬렉션의 원소를 tuple형태로 반환
            num += 1
            excel_sheet.append([num, item['title'], item['link']])
    else:
        print("Error Code:", res.status_code)

excel_file.save('IT_news.xlsx')
excel_file.close()
````
